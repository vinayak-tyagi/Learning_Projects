```python
def plots(ims, figsize = (12,6), rows=1, interp = False, titles = None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1]!=3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    cols = len(ims)//rows if len(ims) % 2 == 0 else len(ims)// rows+1
    for i in range(len(ims)):
        sp = f.add_subplot(rows,cols, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(titles[i], fontsize = 16)
        plt.imshow(ims[i],interpolation=None if interp else 'none') 
```

# Build & train CNN

```python
model = Sequential([ # 1-->no of neuron 2--> kernal in 1st layer
    Conv2D(32,(3,3), activation='relu', input_shape = (224,224,3)),
    Flatten(),
    Dense(2, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
print(train_batches)
model.fit_generator(train_batches, steps_per_epoch=4, validation_data=valid_batches, validation_steps=4, epochs=10, verbose=1)
```

```python
# Savin and loading the model
def saving_model(model_instance,model_name):
    model_instance.save(str(model_name))
    print("Model Saved")
    
def loading_model(model_name):
    model = load_model(str(model_name))
    print("Model Loaded")
    return model

saving_model(model,"CatVsDog.h5")
```
